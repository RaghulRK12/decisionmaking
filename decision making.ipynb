```python
print("Enter Roll number:")
rno = input()
print("Enter student Name:")
sname = input()
print("Enter class:")
cl = input()
print("Enter mark1:")
m1 = int(input())
print("Enter mark2:")
m2 = int(input())
print("Enter mark3:")
m3 = int(input())
print("Enter mark4:")
m4 = int(input())
print("Enter mark5:")
m5 = int(input())

tot = m1 + m2 + m3 + m4 + m5
average = tot / 5
grade = ""

# Determine pass/fail
if m1 > 50 and m2 > 50 and m3 > 50 and m4 > 50 and m5 > 50:
    res = "pass"
else:
    res = "fail"

# Determine grade
if average >= 75:
    grade = "A"
elif 60 <= average < 75:
    grade = "B"
else:
    grade = "C"

# Print results
print("Roll number:", rno)
print("Student name:", sname)
print("Class:", cl)
print("Mark 1:", m1)
print("Mark 2:", m2)
print("Mark 3:", m3)
print("Mark 4:", m4)
print("Mark 5:", m5)
print("Total:", tot)
print("Average:", average)
print("Result:", res)

# Display grade only if the result is "pass"
if res == "pass":
    print("Grade:", grade)
else:
    print("Grade: Not Applicable")
```
